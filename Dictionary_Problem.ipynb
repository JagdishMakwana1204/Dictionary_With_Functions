```python
# Write a function that takes a nested dictionary and prints it in a readable format.

def nested_dict(data, indent=0):     # indent parameter to manage the indentation level
    for key, value in data.items():
        print('  ' * indent + str(key) + " :") # use colon to indicate nested structure
        if isinstance(value, dict):
            nested_dict(value, indent + 1)   # Used recursion to handel nested dictionary
        else:
            print('  ' * (indent + 1) + str(value)) #Used indentation for batter readability


result = nested_dict({
    "Jagdish": {
        "marks": {
            "Maths": 90,
            "Chemistry": 65,
            "Physics": 75
        }
    }
})

print(result)
```

```python
# Count how many times each character appears in a string and store results in a dictionary.
def count_characters(dictionary, string):
    for char in string:
        if char in dictionary:
            dictionary[char] += 1
        else:
            dictionary[char] = 1
    return dictionary

result = count_characters({}, "Hello World")
print(result)
```

```python
# Write a program to invert keys and values of a dictionary.
def invert_dict(origional_dict):
    invert_dict = {}
    for key, value in origional_dict.items():
        invert_dict[value] = key    
    return invert_dict
sample_dict = {'a': 1, 'b': 2, 'c': 3}
result = invert_dict(sample_dict)
print(result)
```

```python
# 11. Write a function that takes a dictionary and returns the sum of all values.
def dict_sum(dictionary, sum):
    for key in dictionary:
        sum += dictionary[key]
    return sum

result = dict_sum({"a" : 45, "b" : 88, "c" : 87}, 0)
print(result)
```

```python
#  Write a function to return the key with the maximum value from a dictionary.
def max_value(dictionary):
    max_val = float('-inf')
    for key in dictionary:
        if dictionary[key] > max_val:
            max_val = dictionary[key]
    return max_val

result = max_value({'a' : 67, 'b' : 78, 'c': 65})
print(result)

# Write a function to return the key with the maximum value from a dictionary.
# def max_value(key):
#     return max(key, key=key.get)  

# result = max_value({'a' : 67, 'b' : 78, 'c': 65})
# print(result)   
```

```python
# 15. Write a function that counts word frequency from a sentence using a dictionary.
def word_count(sentence):
    words = sentence.split()          #First split the sentence into the words
    new = {}                           #used for the dictionary for store the words 
    for words in sentence:
        if words in new:
            new[words] += 1
        else:
            new[words] = 1
    return new
result = word_count("This is Jagdish Makwana")
print(result)
```

```python
# Write a function that takes a dictionary of students & marks, and returns students who scored above 50.
def student_marks(dictionary):
    new = {}
    for student, marks in dictionary.items():
        if marks > 75:
            new[student] = marks
    return new
result = student_marks({"a": 45, "b": 88, "c": 87})
print(result)
```

```python
def swapp_keys_values(dictionary):
    new = {}
    for key in dictionary:
        new[dictionary[key]] = key     # Used for the swapping keys and values
    return new

result = swapp_keys_values({'a' : 45 , 'b' : 57, 'c' : 567})
print(result)
```

```python
# Write a function to multiply all values in a dictionary.

def same_dict(dict1, dict2):

    if dict1.keys() != dict2.keys():    # Used to compare keys of both dictionary
        return False
    for key in dict1:
        if dict1[key] != dict2[key]:    # Used to compare values of both dictionary
            return False
    return True

result = same_dict({'a' :1, 'b' : 3}, {'a' : 1, 'b' : 2})
print(result)                       # Output : False bcs values are different

result = same_dict({'a' :1, 'b' : 2}, {'a' : 1, 'b' : 2})
print(result)                       # Output : True bcs Both dictionary are same
```

```python
# Write a function to multiply all values in a dictionary.

def multiply(dict1, dict2):
    result = {}
    for key in dict1:
        if key in dict2:
            result[key] = dict1[key] * dict2[key]  # Used for the multiplication of values
#      return result                                # only returns first kw multiplication
    return result                                   # correct placement to muultiply all keys  
    
    
result2 = multiply({'a' : 2, 'b' : 4}, {'a' : 6, 'b' : 8})
print(result2)
```
